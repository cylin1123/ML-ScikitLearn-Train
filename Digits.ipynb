```python
from sklearn import datasets
import numpy as np

digits = datasets.load_digits()
    
# Get the keys of the `digits` data
print digits.keys()

# Print out the data
print digits.data
print digits.data.shape #可得資料有1797個觀測值，64 個變數

# Print out the target values
print digits.target
print digits.target.shape #可得資料有1797個目標值（標籤值）
print len(np.unique(digits.target)) #可得知所有的目標值只有 10 個相異值：0 到 9

# Print out the description of the `digits` data
print digits.DESCR
```

```python
print digits.images.shape # 三個維度：1797 個 8 x 8 像素的矩陣

# 將 digits.images 轉換（reshape）為兩個維度，並且使用 numpy 的 all() 方法比較陣列內的元素是否與 digits.data 完全相同
print np.all(digits.images.reshape(1797,64) == digits.data)
```

```python
import matplotlib.pyplot as plt

# 設定圖形的大小（寬, 高）
fig = plt.figure(figsize=(4, 2)) #設定一個長 2 吋，寬 4 吋的空白畫布，準備待會將子圖形畫在上面

fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# 把前 8 個手寫數字顯示在子圖形, 使用一個 for 迴圈開始要將空白畫布填滿
for i in range(8):  
    # 在 2 x 4 網格中第 i + 1 個位置繪製子圖形，並且關掉座標軸刻度
    ax = fig.add_subplot(2, 4, i + 1, xticks = [], yticks = [])
    # 顯示圖形，色彩選擇灰階
    ax.imshow(digits.images[i], cmap = plt.cm.binary)
    # 在左下角標示目標值
    ax.text(0, 7, str(digits.target[i]))

# 顯示圖形
plt.show()
```

```python
from sklearn import datasets
# 載入 matplotlib
import matplotlib.pyplot as plt

# 載入 `digits`
digits = datasets.load_digits()

# 將觀測值與目標值放入一個 list
images_and_labels = list(zip(digits.images, digits.target)) #將兩個陣列存入 images_and_labels變數

# list 中的每個元素
for i, (image, label) in enumerate(images_and_labels[:8]):
    # 在 i + 1 的位置初始化子圖形
    plt.subplot(2, 4, i + 1)
    # 關掉子圖形座標軸刻度
    plt.axis('off')
    # 顯示圖形，色彩選擇灰階
    plt.imshow(image, cmap = plt.cm.binary)
    # 加入子圖形的標題
    plt.title('Training: ' + str(label))

# 顯示圖形
plt.show()
```
